```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_original_img=cv2.imread("./binary.png",0)
```

```python
my_kernel=np.ones((5,5),dtype=np.uint8)
my_kernel
```

```python
my_erosion_img =cv2.erode(my_original_img,my_kernel,iterations=2)

fig=plt.figure(figsize =(10,7))
rows=1
columbs=2


fig.add_subplot(rows,columbs,1)
plt.imshow(my_original_img,cmap="gray")
plt.axis('off')
plt.title("Orginal Image")

fig.add_subplot(rows,columbs,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis('off')
plt.title("Ero Image")
```

```python
my_dialtion_img =cv2.dilate(my_original_img,my_kernel,iterations=2)

fig=plt.figure(figsize =(10,7))
rows=2
columbs=3


fig.add_subplot(rows,columbs,1)
plt.imshow(my_original_img,cmap="gray")
plt.axis('off')
plt.title("Orginal Image")

fig.add_subplot(rows,columbs,2)
plt.imshow(my_erosion_img,cmap="gray")
plt.axis('off')
plt.title("Ero Image")

fig.add_subplot(rows,columbs,3)
plt.imshow(my_dialtion_img,cmap="gray")
plt.axis('off')
plt.title("Dilate Image")
```

```python
new_img=cv2.imread("./binary_with_text.png",0)
```

```python
white_noise=np.random.randint(0,2,size=new_img.shape[:2])
white_noise=white_noise*255
```

```python
my_noised_img=new_img+white_noise
plt.imshow(my_noised_img,cmap="gray")
```

```python
my_opening_img=cv2.morphologyEx(my_noised_img.astype(np.float32),cv2.MORPH_OPEN,my_kernel)
```

```python
fig=plt.figure(figsize =(10,7))
rows=1
columbs=3


fig.add_subplot(rows,columbs,1)
plt.imshow(new_img,cmap="gray")
plt.axis('off')
plt.title("New Image")

fig.add_subplot(rows,columbs,2)
plt.imshow(my_noised_img,cmap="gray")
plt.axis('off')
plt.title("Nosised Image")

fig.add_subplot(rows,columbs,3)
plt.imshow(my_opening_img,cmap="gray")
plt.axis('off')
plt.title("Opening Image")
```

```python
black_noise=np.random.randint(0,2,size=new_img.shape[:2])
black_noise=black_noise* -255
my_black_noised_img =new_img + black_noise
```

```python
print(np.max(black_noise))
print(np.min(black_noise))
```

```python
plt.imshow(black_noise,cmap="gray")
```

```python
my_black_noised_img[my_black_noised_img<=-245]=0
```

```python
my_closing_img=cv2.morphologyEx(my_black_noised_img.astype(np.float32),cv2.MORPH_CLOSE,my_kernel)
```

```python
fig=plt.figure(figsize =(10,7))
rows=1
columbs=3


fig.add_subplot(rows,columbs,1)
plt.imshow(new_img,cmap="gray")
plt.axis('off')
plt.title("New Image")

fig.add_subplot(rows,columbs,2)
plt.imshow(my_black_noised_img,cmap="gray")
plt.axis('off')
plt.title("Black Nosised Image")

fig.add_subplot(rows,columbs,3)
plt.imshow(my_closing_img,cmap="gray")
plt.axis('off')
plt.title("Closing Image")
```

```python
my_gradient_img_1 =cv2.morphologyEx(my_original_img,cv2.MORPH_GRADIENT,my_kernel)
my_gradient_img_2 =cv2.morphologyEx(new_img,cv2.MORPH_GRADIENT,my_kernel)
```

```python
fig=plt.figure(figsize =(10,7))
rows=2
columbs=2


fig.add_subplot(rows,columbs,1)
plt.imshow(my_gradient_img_1,cmap="gray")
plt.axis('off')
plt.title("Grandient 1")


fig.add_subplot(rows,columbs,2)
plt.imshow(my_gradient_img_2,cmap="gray")
plt.axis('off')
plt.title("Gradient 2 Image")

fig.add_subplot(rows,columbs,3)
plt.imshow(my_original_img,cmap="gray")
plt.axis('off')
plt.title("Orginal Image")


fig.add_subplot(rows,columbs,4)
plt.imshow(new_img,cmap="gray")
plt.axis('off')
plt.title("New Image")
```
